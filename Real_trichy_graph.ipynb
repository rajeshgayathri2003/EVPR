```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Define the place
place_name = "Tiruchirappalli, Tamil Nadu, India"

# Download road network data
G = ox.graph_from_place(place_name, network_type='drive')

# Add edge lengths (distances) as weights
G = ox.distance.add_edge_lengths(G)

# Extract edges with weights
edges_with_weights = [
    {
        "from": u,
        "to": v,
        "distance": data["length"]
    }
    for u, v, data in G.edges(data=True)
]

# Print the list of edges with weights
for edge in edges_with_weights[:10]:  # Display the first 10 edges for brevity
    print(f"From: {edge['from']}, To: {edge['to']}, Distance: {edge['distance']:.2f} meters")

# Total number of edges
print(f"\nTotal edges: {len(edges_with_weights)}")
```

```python
# Define the place
place_name = "Tiruchirappalli, Tamil Nadu, India"

# Download road network data
G = ox.graph_from_place(place_name, network_type='drive')

# Plot the road network
fig, ax = ox.plot_graph(
    G,
    node_size=10,
    node_color="red",
    edge_linewidth=0.8,
    edge_color="blue",
    bgcolor="white",
    figsize=(12, 8)
)

# Save the plot as an image
plt.savefig("trichy_road_network.png", dpi=300, bbox_inches="tight")
plt.show()

# Save the graph to a file
ox.save_graphml(G, "trichy_road_network.graphml")

print("Graph plotted, saved as PNG, and stored in GraphML format.")
```
